##Merge

image.png

image.png

````python

# Sample DataFrames
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value1': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value2': [5, 6, 7, 8]})

# Inner Join (Intersection)
inner_join = pd.merge(df1, df2, on='key', how='inner')
print("Inner Join:\n", inner_join)

# Left Join (All from left, matching from right)
left_join = pd.merge(df1, df2, on='key', how='left')
print("\nLeft Join:\n", left_join)

# Right Join (All from right, matching from left)
right_join = pd.merge(df1, df2, on='key', how='right')
print("\nRight Join:\n", right_join)

# Outer Join (Union of both)
outer_join = pd.merge(df1, df2, on='key', how='outer')
print("\nOuter Join:\n", outer_join)
````

````python



# Sample DataFrames
df1 = pd.DataFrame({'key1': ['A', 'B', 'C', 'D'], 'value1': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key2': ['B', 'D', 'E', 'F'], 'value2': [5, 6, 7, 8]})



# Inner Join (Intersection) on different columns
# Let's assume 'key1' in df1 somehow corresponds to 'key2' in df2
inner_join = pd.merge(df1, df2, left_on='key1', right_on='key2', how='inner')
print("Inner Join:\n", inner_join)



# Left Join (All from left, matching from right)
left_join = pd.merge(df1, df2, left_on='key1', right_on='key2', how='left')
print("\nLeft Join:\n", left_join)



# Right Join (All from right, matching from left)
right_join = pd.merge(df1, df2, left_on='key1', right_on='key2', how='right')
print("\nRight Join:\n", right_join)



# Outer Join (Union of both)
outer_join = pd.merge(df1, df2, left_on='key1', right_on='key2', how='outer')
print("\nOuter Join:\n", outer_join)
````

````python
# Concatenating DataFrames Horizontally


import pandas as pd

df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

df2 = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]})

result = pd.concat([df1, df2], axis=1)

print(result)
````

````python
# Concatenating DataFrames Vertically
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

df2 = pd.DataFrame({'A': [7, 8, 9], 'B': [10, 11, 12]})

result = pd.concat([df1, df2], axis=0)

print(result)
````
